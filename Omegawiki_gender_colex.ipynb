```python
male_dict = data_male.groupby(["hyponym", "lang"]).gloss.unique().to_dict()

synsets_male_lemmas = {}
for (synset, lang),lemma in male_dict.items():
    if synset in fraud:
        continue
    if synset not in synsets_male_lemmas:
        synsets_male_lemmas[synset] = {}
    if lang not in synsets_male_lemmas[synset]:
        synsets_male_lemmas[synset][lang] = []
    synsets_male_lemmas[synset][lang].extend(list(lemma))
    
female_dict = data_female.groupby(["hyponym", "lang"]).gloss.unique().to_dict()

synsets_female_lemmas = {}
for (synset, lang),lemma in female_dict.items():
    if synset in fraud_f:
        continue
    if synset not in synsets_female_lemmas:
        synsets_female_lemmas[synset] = {}
    if lang not in synsets_female_lemmas[synset]:
        synsets_female_lemmas[synset][lang] = []
    synsets_female_lemmas[synset][lang].extend(list(lemma))
```

```python
synsets_colexified = []

for i, synset_f in enumerate(synsets_female_lemmas):
    for j, synset_m in enumerate(synsets_male_lemmas):
        if synset_f != synset_m:
            langs = {}
            match = False
            for lang in synsets_male_lemmas[synset_m]:
                if lang in synsets_female_lemmas[synset_f]:
                    f_set = set(synsets_female_lemmas[synset_f][lang])
                    m_set = set(synsets_male_lemmas[synset_m][lang])
                    
                    colexified_lemmas = m_set & f_set
                    
                    if len(colexified_lemmas) > 0:
                        match = True
                        if lang not in langs:
                            langs[lang] = []
                        langs[lang].extend(colexified_lemmas)
            if match:
                synsets_colexified.append({'synset_f':synset_f, 'synset_m':synset_m, 'colexified_lemmas':langs})
```

```python
colexified_m = []
colexified_f = []
female_nodes = []
for pair in synsets_colexified:
    if pair['synset_f'] not in female_nodes:
        female_nodes.append(pair['synset_f'])
    if pair['synset_f'] not in colexified_f:
        colexified_f.append(pair['synset_f'])
    if pair['synset_m'] not in colexified_m:
        colexified_m.append(pair['synset_m'])
colexified_set = colexified_f + colexified_m
```

```python
synsets_matrix = np.zeros((len(colexified_set),len(colexified_set)))
langs_att = {}
for i, synset1 in enumerate(colexified_set):
    j = i + 1
    for synset2 in colexified_set[j:]:
        if j < len(colexified_set):
            for pair in synsets_colexified:
                if (synset1 == pair['synset_f'] and synset2 == pair['synset_m']) or (synset1 == pair['synset_m'] and synset2 == pair['synset_f']):
                    langs = list(pair['colexified_lemmas'].keys())
                    langs_str = ""
                    if len(langs) > 8:
                        for lang in langs:
                            langs_str += lang[:3] + " "
                    else:
                        langs_str = "    ".join(langs)
                    langs_att[(pair['synset_f'],pair['synset_m'])] = langs_str
                    synsets_matrix[i,j] = len(pair['colexified_lemmas'])
                    synsets_matrix[j,i] = len(pair['colexified_lemmas'])
            j += 1
```

```python
synsets_matrix
```

```python
G = nx.from_numpy_matrix(synsets_matrix)
G = nx.relabel_nodes(G, lambda x: list(colexified_set)[x])
```
